### construct a dataframe from Like list

```python
# go to Likes page

driver.get("https://www.facebook.com/" + fid + "/likes")
```

```python
# auto scroll

current_len = 0
last_len = -1
while current_len != last_len:
    last_len = current_len
    current_len = driver.execute_script("window.scrollTo(0, document.body.scrollHeight); return document.body.scrollHeight;")
    #driver.implicitly_wait(5) # ask browser to wait x seconds
    time.sleep(2) # ask python to wait for x seconds, pages need some time to be loaded
```

```python
html_doc = driver.page_source
html_doc = html_doc.replace("<!-- ", "")
html_doc = html_doc.replace(" -->", "")

soup = BeautifulSoup(html_doc, 'html.parser')

likes = soup.find_all("div", class_= "fsl fwb fcb")
print(len(likes)) # print how many likes
for like in likes:
    href = like.find("a")['href']
    print(like.text, href)
```

### Save likes into a CSV file (can be open by Microsoft EXCEL)

```python
# after you have "likes", save them into a CSV file 
csv_list = list()

for like in likes:
    # data 
    href = like.find("a")['href']
    text = like.text
    
    data = [fid, text, href]
    csv_list.append(data)
```

```python
#print(csv_list)
```

```python
df = pd.DataFrame(csv_list, columns = ['fid', 'name', 'url'])
df # show all data
```

```python
new_dir = new_dir + fid.replace('.', '') + '/'
if not os.path.isdir(new_dir):
    os.makedirs(new_dir)

print(new_dir)
```

```python
df.to_csv(new_dir + "/likes.csv", index = False)
```
