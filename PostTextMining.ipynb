```python
# 2) Read the page.html
with open(new_dir + fid.replace(".", "") + "/" + "post.html", "rb") as in_file:
    html_doc = in_file.read().decode("utf-8")
```

```python
# 3) modify page.html
html_doc = html_doc.replace("<!-- ", "")
html_doc = html_doc.replace(" -->", "")

soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
def href2fId(href):
    fid = href.split("?")[0].split("/")[-1]
    fid = fid if fid != "profile.php" else href.split("id=")[-1].split("&")[0]
    return fid
```

```python
# 4) parse all post into a "set"
post_set = set()
comment_set = set()

# parse: you can see "selenium-fb.ipynb"
contents = soup.find_all("div", class_ = "fbUserStory") 
for content in contents:
    post = content.find("p")
    if post:
        
        # time
        time_human = content.find("abbr")["title"]
        time_utime = content.find("abbr")["data-utime"]

        # number of likes
        likes = content.find("div", class_ = "UFILikeSentence")
        if likes: # sometimes a post may have no 'likes'
            likes = likes.find("a")["aria-label"]
            number_of_likes = int(likes.split(' ')[0])
        else:
            number_of_likes = 0

        # comments
        commentblocks = content.find_all("div", class_ = "UFICommentContentBlock")
        
        #print result if you want
        #print(post.text)
        #print(time_human, time_utime)
        #print("number of likes:", number_of_likes)
        
        #print how many comments if you want
        #print("number of comments:", len(commentblocks))
        
        # MIKE: 20171023, add post into post_set()
        if post.text:
            post_set.add(post.text)
        
        # parse comments
        for commentblock in commentblocks:
            commenter_name = commentblock.find("a",  class_ = "UFICommentActorName").text
            commenter_href = commentblock.find("a",  class_ = "UFICommentActorName")['href']
            commenter_fid = href2fId(commenter_href)
            comment = commentblock.find("span",  class_ = "UFICommentBody").text
            comment_time_human = commentblock.find("abbr")["title"]
            comment_time_utime = commentblock.find("abbr")["data-utime"]
            
            #print each comment data if you want
            #print(commenter_name, commenter_fid)
            #print(comment)
            #print(comment_time_human, comment_time_utime)
            
            # MIKE: 20171023, add comments into comment_set()
            if comment:
                comment_set.add(comment)
        
    #print(content)
    #print("======")
```

### 無字庫斷詞斷字

```python
import string
pun_set = {'。', ' ', '，', '「', '」', '？', '！', '（', '）', '：', '…', '～', '、'}
pun_set = pun_set.union(set(string.punctuation)) # english punctuation
#print(pun_set)
```

```python
def allPrintable(word):
    for c in word:
        if c not in string.printable:
            return False
    return True
```

```python
import jieba
import jieba.analyse
```

```python
post_word_list = list()

jieba.set_dictionary('dict.txt.big') # 繁體中文設定
jieba.analyse.set_stop_words("stop_words.txt") # 去除 stop words

for sentence in post_set:
    words = jieba.cut(sentence, cut_all=False)
    for word in words:
        if len(word) >= 1 and word and (word not in pun_set) and not allPrintable(word): # 去除小於一的字元的字
            post_word_list.append(word)
```
